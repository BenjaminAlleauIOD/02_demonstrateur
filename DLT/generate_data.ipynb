```python
import pandas as pd
import numpy as np
import datetime

# Nombre de CADs
num_cads = 15

# Nombre de lignes par CAD
lines_per_cad = 20

# Créer un dataframe vide
data = []

dimensions = [160,180, 200, 220]

# Remplir le dataframe
for i in range(num_cads):
    cad = f"001479_{101+i}"
    dimension = np.random.choice(dimensions)
    dimensions.remove(dimension)  # Retirer la dimension choisie de la liste pour éviter les répétitions
    if not dimensions:
        dimensions = [160,180, 200, 220]  # Réinitialiser la liste si toutes les dimensions ont été utilisées
    
    for _ in range(lines_per_cad):
        
        if dimension == 180:
            dlt = int(np.random.normal(25, 5))
        elif dimension == 200:
            dlt = int(np.random.normal(20, 5))
        else:
            dlt = int(np.random.normal(30, 10))
        
        start_date = datetime.date(2023, 1, 1)
        end_date = datetime.date(2023, 6, 30)
        delta = end_date - start_date
        random_days = np.random.randint(delta.days)
        date = start_date + datetime.timedelta(days=random_days)
        
        data.append([cad, dimension, date, dlt])

df = pd.DataFrame(data, columns=["CAD", "Dimension", "Date", "DLT"])
```
